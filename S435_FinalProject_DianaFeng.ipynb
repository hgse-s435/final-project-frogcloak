# Introduction

## Premise
 * Development of mobile, affordable sensors makes it possible to study natural social behaviors with an objective acuity never before seen in the social sciences
 * Social interactions are believed to be particularly influential in teaching and learning, particularly if viewed from a socio-constructivist perspective
 * The current study made use of mobile eye-trackers, physiological sensors, and voice and movement sensors to study learner behavior in a collaborative learning task

## Study
* Setup: 
Pairs (n = 42) of collaborators trying to solve a maze using the GoGo Board
* Design:
2 by 2 design with
   * One group receiving both verbal explanation of the benefits of collaboration and real-time visualization of each    * partner’s speech ratio during the session
   * One group receiving only the verbal explanation
   * One group receiving only visualization
   * The last group receiving no intervention

## Dataset
 * Sensor data:
    * Gaze information from mobile eye trackers (Tobii Glasses 2)
    * Physiological response (e.g. heart rate, electrodermal activity, skin temperature) from Empatica E4 wristbands
    * Speech and body movement from Kinect sensors
 * Questionnaire data:
    * Pre- and post- learning tests
    * Self evaluation of collaboration quality based on task division, time management, reciprocal interaction, reaching consensus, dialog management, and sustaining mutual understanding
    * Researcher’s assessment of the collaboration with similar measures, as well as the quality of task completion


# Data Pre-Cleaning (Out of Python)

## Preliminary observations:
* After dropping missing sessions from all three data streams (Gaze, EDA and Speech), only 31 sessions are left for analysis
* Participants in each pair can be matched with "x" and "y"
* EDA data most likely need to be manually trimmed to sync with the gaze data by checking the tags for each pair
  * Since the tags are not all consistent for each file, it is best to trim the csvs themselves individually instead of trimming the data after entering them into python


## Existing analysis:
* Gaze
    * Joint Visual Attention was calculated prior by first matching the gaze point to a ground truth through the use of fiducial markers, and then calculate by assessing whether the two participants’ gazes were within certain amounts of pixels within a 4-second time frame

* Physiological data
    * The synchrony of electrodermal activity (EDA) was analyzed within each pair through different modes of measurements and correlated to self- and researcher-ratings of the collaboration and task performance

* Kinect data
    * The movement of various joints were analyzed and clustered to study the body postures and positions correlated to good collaboration

## Data cleaning performed out of python:
1. EDA data: trimmed the recording session based on tags and gaze recording timestamps (performed manually for each file)
2. Speech data: (performed using an Excel file with written functions)
    * Converted start and end times to frames
    * Substracted the frame number by the frame number of the beginning of the gaze segment
    * Removed any entries that begin or end out side of the gaze segment
3. Building data folder: organized data by sessions, with each folder containing a cleaned EDA file, two cleaned speech files (one for each participant in the group), and one gaze file

### Brief summary
 * P values:
     * JVA and left speech: 0.22382281582796867 > 0.05, not correlated
     * JVA and right speech: 6.119035117816121e-11 << 0.05, there is a correlation. Comparing the observed values with expected values, we find that the co-occurrence of JVA and right speech is lower than expected. So is the simultaneous absence of the two parameters.
     * JVA and simulaneous speaking: 0.3315059996533162 > 0.05, not correlated
     * JVA and either speech: 1.2505333491646183e-11 << 0.05, there is a correlation. Comparing the observed values with expected values, we find that the co-occurrence of JVA and either speech is lower than expected. So is the simultaneous absence of the two parameters.

 * Conclusion:
      * There appears to be some interaction between JVA and speech occurrence. Specifically, there appears to be a negative correlation between JVA and either of the participants speaking at the same time

 * Limitation:
      * Sphinx algorithm cannot accurately determine if one person is speaking versus the other. The most realiable speech data is limited to cases where either participant was speaking
      * Due to missing data, only 38.8% of the data could be preserved for correlational study, which significantly limited the validity of the result

### Step 2: Choosing cluster size using Kmeans

```python
from sklearn.cluster import KMeans

ks = list(range(1, 10))
inertias = []

for k in ks:
    kmeans = KMeans(n_clusters=k, max_iter=1000)
    kmeans.fit(cluster_df.values)
    inertias.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

It appears that the rate of change of the inertia becomes rather stable after cluster number of 5. In the following analysis, I will compare the results from a cluster size of 3 (same as published paper) and 5.

Ideally, I would have run a T-SNE analysis to visualize the clusters, but since there are more than 230k data points, this process takes an extraordinarily long time to complete. I waited for 30 minutes when I tested it and had to abort before it could finish.
